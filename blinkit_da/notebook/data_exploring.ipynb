```python
#function for get columsn datatypes 
def get_datatypes(df):
    '''function for get datatypes'''
    categorical=df.select_dtypes(include=['object','category']).columns.tolist()
    numeric=df.select_dtypes(include=['number']).columns.tolist()
    return categorical,numeric
```

# visulizaion


** chart requirments **

```python
#total_sales by fatcontent
sales_by_fat=df.groupby("Item Fat Content")['Sales'].sum()
sales_by_fat

plt.pie(sales_by_fat,labels=sales_by_fat.index,
        autopct="%.1f%%",
        shadow=True,
        
        
        startangle=90)
plt.title('sales by fat')
plt.axis('equal')
plt.show()
```

```python
#total_sales by item type 
sales_by_item=df.groupby("Item Type")['Sales'].sum().sort_values(ascending=False)
sales_by_item
plt.figure(figsize=(13,7))

bars=plt.bar(sales_by_item.index,sales_by_item.values)

#to writen the text on bar 
for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(),
             f'{bar.get_height():,.0f}',ha='center',va='bottom',fontsize=8)
plt.xticks(rotation=91)
plt.title('sales by item')
plt.xlabel('item')
plt.ylabel('sales')
plt.tight_layout()
plt.show()
```
